```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("DK- NALSA Remand & Bail Responses.xlsx")
```

```python
df.head()
```

```python
districts = df["Q.No."].values
```

```python
questions = df.T["State"].values
```

```python
tdf = df.T
```

```python
tdf.index
```

```python
tdf.columns = tdf.columns.str.strip()
tdf.columns = pd.MultiIndex.from_tuples(zip(tdf.columns, districts))
tdf.index = tdf["State"].values
tdf.columns
```

```python
tdf= tdf[(tdf["State"] != "District").values]
```

```python
df
```

```python
df = tdf.T
```

```python
df.columns
```

```python
df.head()
```

```python
df = df.replace(to_replace="Not Responded", value="NR")
df = df.replace(to_replace="Not responded", value="NR")
df = df.replace(to_replace="Not provided", value="NP")
df = df.replace(to_replace="Not Provided", value="NP")
df = df.replace(to_replace="Attachment not found", value="NP")
df = df.replace(to_replace="To check list", value="NP")
df = df.replace(to_replace="Yes", value="YES")
df = df.replace(to_replace="yes", value="YES")
df = df.replace(to_replace="No", value="NO")
df = df.replace(to_replace="no", value="NO")
df = df.replace(to_replace="No fixed tenure", value=-1)
df = df.replace(to_replace="No, but PLVs have been appointed", value="NO")
df = df.replace(to_replace="Maintained by the Police stations", value="NP")
```

```python
df = df.fillna(value="NA")
```

```python
df
```

```python
df=df[1:]
```

```python
colnames =  ['LEGAL_AID_MANDATORY','NUM_MAGISTRATE_COURTS_TOTAL','NUM_MAGISTRATE_COURTS_REMAND',
             'NUM_REMAND_BAIL_LAWYERS','TENURE_LWYRS','LWYRS_APNTD_FOR_REMAND_COURTS','LWYRS_TRAINED','NUM_ACCSD_RPRSNTD_POLSTATN',
             'NUM_ACCSD_RPRSNTD_COURT','NUM_GRANTED_BAIL_POLSTATN','NUM_GRANTED_BAIL_COURT','NUM_RELEASED_POLSTATN','NUM_RELEASED_COURT',
             'NUM_DISCHARGED','DLSA_TO_SLSA','NUM_ATTNDNC_CERTIFICATE','NUM_REPORTS_LWYRS_SUBMITTED','NUM_CASES_FEE_COMPLAINT','NUM_LWYRS_REMOVED']
df.columns = colnames
```

```python
df.to_excel("remand_and_bail_cleaned.xls")
```

```python
final_list = []
for value in pd.Series(df.index.get_level_values(0)).unique():
    state_dict = {}
    for k,v in df.T[value].to_dict().iteritems():
        state_dict["STATE"] = str(value)
        state_dict["DISTRICT"] = str(k)
        state_dict.update(v)
    final_list.append(state_dict)
print final_list
```
